## 2\. Números

Preencha as células de código para preencher os valores de (A), (B) e (C) na tabela de ticket médio abaixo:

<br>

| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio |
|-------|--------------------|------------------|--------------|
| 19/01 | (A)                | 3                | 320.52       |
| 20/01 | 834.47             | (B)              | 119.21       |
| 23/01 | 15378.12           | 5                | (C)          |

```
# (A)
valor_vendas_a = 0
qtd_vendas_a = 3
ticket_medio_a = 320.52
valor_vendas_a = qtd_vendas_a * ticket_medio_a
print(valor_vendas_a)
```

```
# (B)
valor_vendas_b = 834.47
qtd_vendas_b = 0
ticket_medio_b = 119.21
qtd_vendas_b = valor_vendas_b // ticket_medio_b
print(qtd_vendas_b)
```

```
# (C)
valor_vendas_c = 15378.12
qtd_vendas_c = 5
ticket_medio_c = 0
ticket_medio_c = valor_vendas_c / qtd_vendas_c
print(ticket_medio_c)
```
